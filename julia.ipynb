```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def newton(z0, max_iter=100, acc=1e-10):
    z = z0
    for i in range(max_iter):
        fz = f(z)
        if abs(fz) < acc:
            return i, z
        dz = df(z)
        if dz == 0:
            break
        z = z - fz / dz
    return max_iter, z
```

```python
xmin, xmax = -2, 2
ymin, ymax = -2, 2
num_points = 5000
x = np.linspace(xmin, xmax, num_points)
y = np.linspace(ymin, ymax, num_points)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y
c = complex(-0.8, 0.156)
```

```python
julia = np.zeros(Z.shape)

limit = 100

for i in range(num_points):
    for j in range(num_points):
        z = Z[i, j]
        for k in range(limit):
            z = z**2 + c
            R = (1 + math.sqrt(1+4*abs(c)))/2
            if abs(z) > R:
                julia[i, j] = k
                break

```

```python
plt.figure(figsize=(12, 12))
plt.imshow(julia, cmap='inferno', origin='lower')
plt.colorbar(label='Скорость расхождения')
plt.title('Множество Жюлиа')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

```python
def plot_julia(c):
    julia = np.zeros(Z.shape)
    
    limit = 100
    
    for i in range(num_points):
        for j in range(num_points):
            z = Z[i, j]
            for k in range(limit):
                z = z**2 + c
                R = (1 + math.sqrt(1+4*abs(c)))/2
                if abs(z) > R:
                    julia[i, j] = k
                    break
                    
    plt.figure(figsize=(12, 12))
    plt.imshow(julia, cmap='inferno', origin='lower')
    plt.colorbar(label='Скорость расхождения')
    plt.title('Множество Жюлиа')
    plt.xlabel('Re(z)')
    plt.ylabel('Im(z)')
    plt.show()
```

```python
c_values = [complex(-0.8, 0.156), complex(0.285, 0.01), complex(-0.0085, 0.71)]
```

```python
for c in c_values:
    plot_julia(c)
```
